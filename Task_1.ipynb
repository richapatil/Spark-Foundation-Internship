### Simple Linear Regression

### Question: Predict the percentage of an student based on the no. of study hours

### Final Output to be Predicted: What will be predicted score if a student studies for 9.25 hrs/ day? 

```python
#importing all the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#importing the dataset
data = pd.read_csv("C:/Users/richa/Spark_Foundation/student_scores - student_scores.csv")
data.head(10)
```

```python
data.shape
```

```python
data.describe()
```

```python
#plotting the dataset
data.plot(x = "Hours", y = "Scores", style = "*")
plt.title("Hours vs Percentage")
plt.xlabel("Hours of Study")
plt.ylabel("Percentage Score")
plt.show()
```

```python
#Dividing the given data into "attributes" and "labels" i.e input and output
#Extracting x and y values from the dataset
X = data.iloc[:,:-1].values
y = data.iloc[:,1].values
```

```python
#Splitting dataset into training and testing set.
#For splitting the dataset we need to import the train_test_split() method from sklearn
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, 
                                                    random_state = 0)
```

### Training the algorithm
As we have split the dataset into training and testing, so now we will train our algorithm

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
print("Training has been completed!!")
```

```python
#Plotting the regression line
line = regressor.coef_*X+regressor.intercept_

#Plotting for test data
plt.scatter(X, y, color='red')
plt.plot(X, line, color='black')
plt.show()
```

### Making Prediction

As we have trained the algorithm, now we will try to make some prediction

```python
# Testing data - In Hours
print(X_test) 

# Predicting the scores
y_pred = regressor.predict(X_test) 
```

```python
#Compairing actual and predicted values
df = pd.DataFrame({'Actual': y_test, 'Predicted':y_pred})
df
```

```python
#testing with sample values
hours = [[9.25]]
pred = regressor.predict(hours)
print("If a student studies for {} hours per day he/she will score {}% in exam".format(hours, pred[0]))
```

### Evaluating the Model

```python
from sklearn import metrics
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```
